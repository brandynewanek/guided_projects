```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.svm import NuSVR

import shap
import optuna
```

```python
features = ['Dystopia_1.83_residual', 'GDP_per_capita',
       'Social_support', 'Healthy_life_expectancy',
       'Freedom_of_choices', 'Generosity', 'Perceptions_corruption']
target = ['Happiness_score']
```

```python
def shap_summary_plot(df, features=features, target=target, test_size=.15):
    X = df[features].iloc[14:]
    y = df[target].iloc[14:]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    std_scaler = StandardScaler().fit(X_train)
    X_train = std_scaler.transform(X_train)
    X_test = std_scaler.transform(X_test)
    X_train = pd.DataFrame(X_train, columns=X.columns)
    X_test = pd.DataFrame(X_test, columns=X.columns)

    X_train_summary = shap.kmeans(X_train, 10)

    pca = PCA()
    nsvr = NuSVR()
    pipe = make_pipeline(pca, nsvr).fit(X_train,y_train)

    ex = shap.KernelExplainer(pipe.predict, X_train_summary)
    shap_values = ex.shap_values(X_test)

    shap.initjs()
    
    print('test_score: {}'.format(pipe.score(X_test, y_test)))

    return shap.summary_plot(shap_values, X_test)
    
```
